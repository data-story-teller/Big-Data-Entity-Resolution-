```python
# importing essestial libraries
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import regexp_tokenize
from itertools import combinations
```

```python
# load the datasets: First Entity Collection
data1 = pd.read_csv("Entity_Collection_1.csv")
data1.head()
```

```python
# load the datasets: Second Entity Collection
data2 = pd.read_csv("Entity_Collection_2.csv")
data2.head()
```

## Preprocessing
#### Removing unwanted characters such as hyphen (-), fullstops (.) from the different columns of the data.

```python
# Name Column of entity collection one
data1['name']= data1['name'].str.replace(r"\b(?:[A-Z]\.)+(?!\w)","",regex=True)
data1.tail(10)
```

```python
# full_name Column of entity collection two
data2['full_name']= data2['full_name'].str.replace(r"\b(?:[A-Z]\.)+(?!\w)","",regex=True)
data2.tail(10)
```

```python
# Phone Column of entity collection one
# Removing hyphen and country codes (Make every phone number atmost 10 digits)

data1['phone'] = data1['phone'].str.replace(r'-','',regex=True)
data1['phone'] = data1['phone'].str.replace(r'(','',regex=True)
data1['phone'] = data1['phone'].str.replace(r')','',regex=True)

data1['phone'] = data1['phone'].str[-10:]

data1.tail(10)
```

```python
# phone_num column of entity collection two
# Removing hyphen and country codes (Make every phone number atmost 10 digits)
data2['phone_num'] = data2['phone_num'].str.replace(r'-','',regex=True)
data2['phone_num'] = data2['phone_num'].str.replace(r'(','',regex=True)
data2['phone_num'] = data2['phone_num'].str.replace(r')','',regex=True)

data2['phone_num'] = data2['phone_num'].str[-10:]

data2.tail(10)
```

```python
# University column of entity colection one
# Removing commas (,) and hyphen (-)

data1['university'] = data1['university'].str.replace(r",","", regex=True)
data1['university'] = data1['university'].str.replace(r" - "," ", regex=True)
data1.tail(10)
```

```python
# School column of entity collection two
# Removing commas (,) and hyphen (-)

data2['school'] = data2['school'].str.replace(r",","",regex=True)
data2.tail(10)
```

## Clustering Attributes

```python
# Empty list to store the attributes of both the entity collections
attributes_e1 = []
attributes_e2 = []

for attribute in data1.columns:
    if attribute not in attributes_e1:
        attributes_e1.append(attribute)

for attribute in data2.columns:
    if attribute not in attributes_e2:
        attributes_e2.append(attribute)
```

```python
#attribute names of the first entity collection
attributes_e1
```

```python
#attribute names of the second entity collection
attributes_e2
```

```python
# Generating the token values for each attribute, this values will be used for computing Jaccard Similarities among them 
# So that we can decide which attributes are similar to each other

values_attributes_e1 = {}
values_attributes_e2 = {}

# Checking for stop words removal as they can not be a token or block (Example : 'of' can not be a block)
sw_nltk = stopwords.words('english')
# Checking articles
articles = ['a','A','an','An','AN','the', 'The','THE']


for att in attributes_e1[1:]:
    
    temp_values_e1 =[]
    
    data1[att] = data1[att].str.strip()
    temp_uniques = data1[att].unique()
    temp_text = ' '.join(temp_uniques)
    temp_token = regexp_tokenize(temp_text, "[\w']+")
    
    for i in temp_token:
        if i not in temp_values_e1:
            if i not in sw_nltk:
                if i not in articles:
                    if len(i) >=2:
                        temp_values_e1.append(i)
            
            
    values_attributes_e1[att] = temp_values_e1


for att in attributes_e2[1:]:
    
    temp_values_e2 =[]
    
    data2[att] = data2[att].str.strip()
    temp_uniques = data2[att].unique()
    temp_text = ' '.join(temp_uniques)
    temp_token = regexp_tokenize(temp_text, "[\w']+")
    
    for i in temp_token:
        if i not in temp_values_e2:
            if i not in sw_nltk:
                if i not in articles:
                    if len(i) >=2:
                        temp_values_e2.append(i)
            
            
    values_attributes_e2[att] = temp_values_e2
```

### Jaccard Similarity Function

```python
def jaccard_similarity(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(set(list1)) + len(set(list2))) - intersection
    return float(intersection) / union
```

```python
# Jaccard Similarity function calculation for each attributes of the first entity collections

j_similarity_e1 = {}
for k1,v1 in values_attributes_e1.items():
    temp ={}
    for k2,v2 in values_attributes_e2.items():        
        temp [k2] = jaccard_similarity(v1,v2)

    j_similarity_e1 [k1] = temp
```

```python
j_similarity_e1
```

```python
# creating tuples of highly similar attributes for the attributes of the first entity commection 
l1 = []
for k,v in j_similarity_e1.items():
    for m,n in v.items():
        if n == max(v.values()):
            #print(m, n)
            l1.append((k,m))
        
```

```python
l1
```

```python
# Jaccard Similarity function calculation for each attributes of the second entity collections

j_similarity_e2 = {}
for k1,v1 in values_attributes_e2.items():
    temp ={}
    for k2,v2 in values_attributes_e1.items():        
        temp [k2] = jaccard_similarity(v1,v2)

    j_similarity_e2 [k1] = temp
```

```python
j_similarity_e2
```

```python
# creating tuples of highly similar attributes for the attributes of the first entity commection 

l2 =[]
for k,v in j_similarity_e2.items():
    for m,n in v.items():
        if n == max(v.values()):
            #print(k+","+m, n)
            l2.append((k,m))
```

```python
l2
```

```python
# merge those tuples to cleate a list of attibute tuples which are similar in both entity collection

final_list =[]
all_att = attributes_e1+attributes_e2
for i, j in combinations(all_att, 2):
    temp_list =[]
    if (i,j) in l1 or (i,j) in l2:
        if len((i,j)) != 0:
            temp_list.append((i,j))
    if (j,i) in l1 or (j,i) in l2:
        if len((j,i)) != 0:
            temp_list.append((j,i))
    if len(temp_list) !=0:
        final_list.append((temp_list))
```

```python
final_list
```

```python
# Merge those attribute tuples to cleate clusters
clusters = []
for i in range(len(final_list)):
    temp_cluster =[]
    tup = final_list[i][0]+final_list[i][1]
    for v in tup:
        if v not in temp_cluster:
            temp_cluster.append(v)
            
    clusters.append(temp_cluster)
```

```python
clusters
```

### Token Generation (Blocks)

```python
# Generating blocks for each clusters
clusters_blocks = []

for c in range(len(clusters)):
    temp_token =[]
    for cluster in clusters[c]:
        if cluster in data1.columns:
            for i in range(len(data1)):
                temp_list = regexp_tokenize(data1[cluster][i], "[\w']+")
                
                for j in temp_list:
                    #token = 'c_'+str(c+1)+"."+j
                    if j not in temp_token:
                            if j not in sw_nltk:
                                if j not in articles:
                                    if len(j) >=2:
                                        temp_token.append(j)
                   
                
        elif cluster in data2.columns:
            for i in range(len(data2)):
                temp_list = regexp_tokenize(data2[cluster][i], "[\w']+")
            
                for j in temp_list:
                    #token = 'c_'+str(c+1)+"."+j
                    if j not in temp_token:
                            if j not in sw_nltk:
                                if j not in articles:
                                    if len(j) >=2:
                                        temp_token.append(j)
                  
    clusters_blocks.append(temp_token)                               
    #clusters_tokens['c_'+str(c+1)]= temp_token
        
```

```python
clusters_blocks
```

```python
# empty dictionary
blocks_with_entity = {}

# iterating both entity collections to assign entities to each blocks. 
for m in range(len(clusters_blocks)):
    for block in clusters_blocks[m]:
        temp_list = []
        for i in range(len(data1)):
            for x in clusters[m]:
                if x in attributes_e1:
                    if block in data1[x][i]:
                        temp_list.append(data1['id'][i])
            
        for j in range(len(data2)):
            for x in clusters[m]:
                if x in attributes_e2:
                    if block in data2[x][j]:
                        temp_list.append(data2['id_num'][j])
     
        blocks_with_entity['c'+str(m+1)+"."+block] = temp_list
    
```
